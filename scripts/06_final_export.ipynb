```python
import os
import pandas as pd
 
expected = [
    '../results/daily_annual_max_24h_1950_2024.csv',
    '../results/hourly_annual_max_1h_12h_2000_2024.csv',
    '../results/design_rainfall_gumbel.csv',
    '../results/design_rainfall_gev.csv',
    '../results/idf_intensity_gev.csv',
    '../results/ddf_depth_gev.csv',
    '../results/ks_test_summary.csv',
    '../results/split_period_comparison.csv',
    '../results/probability_plots_all_durations.png',
    '../results/idf_curves.png',
    '../results/ddf_curves.png',
    '../results/idf_ddf_combined.png',
    '../results/trend_analysis.png',
    '../results/monthly_distribution.png',
]
 
print("Checking all result files:")
all_ok = True
for f in expected:
    exists = os.path.exists(f)
    print(f"  {'OK' if exists else 'MISSING':<8} {f}")
    if not exists:
        all_ok = False
 
if all_ok:
    print("\nAll files present — ready for GitHub!")
else:
    print("\nSome files missing — re-run the notebooks above.")
```

```python
gev_depth     = pd.read_csv('../results/design_rainfall_gev.csv',
                             index_col='duration_h').astype(float)
gev_intensity = pd.read_csv('../results/idf_intensity_gev.csv',
                             index_col='duration_h').astype(float)
 
dur_labels = [f'{int(d)}h' for d in gev_depth.index]
gev_depth.index     = dur_labels
gev_intensity.index = dur_labels
 
gev_depth.columns     = [f'T{c}yr_depth_mm'     for c in gev_depth.columns]
gev_intensity.columns = [f'T{c}yr_intensity_mmh' for c in gev_intensity.columns]
 
final_table = pd.concat([gev_depth, gev_intensity], axis=1)
final_table.index.name = 'duration'
final_table.to_csv('../results/idf_table_final.csv')
 
print("Final IDF table (GEV) — Paris-Montsouris:")
print(final_table.round(1).to_string())
print("\nSaved: results/idf_table_final.csv")
```
